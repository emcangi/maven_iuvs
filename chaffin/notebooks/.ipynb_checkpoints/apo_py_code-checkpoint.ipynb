```python
def filetag(fname): #get tag for files, omitting directiory and version/revision number
    return os.path.basename(fname).split("_v")[0]
```

```python
def groupfiles(fnames): #group files by unique tags
    uniquenames=set([filetag(f) for f in fnames])
    return [[f for f in fnames if filetag(f)==u] for u in uniquenames]
```

```python
def fileversionpreference(fname): #gen an integer to sort files in ascending order by recent
    fileversion=fname.split("_v")[-1]
    versionint=int(fileversion[:2])*1000+int(fileversion[4:6])
    if fileversion[3] == "s":
        versionint+=100
    return versionint
```

```python
def fileorder(fname,filelist): # get order according to original list
    return [n for n,f in enumerate(filelist) if f==fname][0]
```

```python
def latestorbfiles(orbfiles):
    return sorted([sorted(l,key=fileversionpreference)[-1] for l in groupfiles(orbfiles)],
                  key=lambda x: fileorder(x,orbfiles))
```

```python
def unit_vector(vector):
    """ Returns the unit vector of the vect?or.  """
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    """ Returns the angle in radians between vectors 'v1' and 'v2'::

            >>> angle_between((1, 0, 0), (0, 1, 0))
            1.5707963267948966
            >>> angle_between((1, 0, 0), (1, 0, 0))
            0.0
            >>> angle_between((1, 0, 0), (-1, 0, 0))
            3.141592653589793
    """
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    return 180/np.pi*np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
```

```python
def lrl_vec(r,v): #needs position and velocity in inertial frame, km, km/s
    vv=1e3*v
    rr=1e3*r
    G=6.67408e-11
    M=6.4171e23
    lang=np.cross(rr,vv)
    lrlvec=np.cross(vv,lang)-G*M*(rr/np.linalg.norm(rr))
    return lrlvec/np.linalg.norm(lrlvec)
```

```python
def getcoordmats(fits,frame):
    if len(frame) > 0:
        frame="_"+frame
    return np.transpose(np.array(
            [fits['SpacecraftGeometry'].data[:]['VX_SPACECRAFT'+frame],
             fits['SpacecraftGeometry'].data[:]['VY_SPACECRAFT'+frame],
             fits['SpacecraftGeometry'].data[:]['VZ_SPACECRAFT'+frame]]),axes=[1,2,0])
```

```python
def gettmats(fits,fromframe,toframe):
    return np.array([np.matmul(tomat,np.linalg.inv(frommat)) for tomat,frommat in zip(getcoordmats(fits,toframe),
                                                                                     getcoordmats(fits,fromframe))])
```

```python
def xyoffset(pixelvec,mat,r,x,y):# need km, km/s
    pv=np.matmul(mat,pixelvec)
    ps=-r-np.dot(r,pv)*pv
    return [np.dot(ps,x),np.dot(ps,y)]
```

```python
def boundarypoly(pixeloffsets):
    polycoords=np.concatenate((pixeloffsets[:,0,0,:],
                               pixeloffsets[-1,:,2,:],
                               pixeloffsets[-1::-1,-1,3,:],
                               pixeloffsets[0,-1::-1,1,:]))
    return polycoords
```

```python
def getcounts(fits):
    l0=121.6
    w=2.5
    counts=np.array([[np.sum(b*(np.abs(wav-l0)<w)) for b,wav in zip(integration,fits['Observation'].data[0]['WAVELENGTH'])] for integration in fits['detector_dark_subtracted'].data])
    cal=np.interp(l0,fits['Observation'].data[0]['WAVELENGTH'][0],fits['Primary'].data[0][0]/fits['detector_dark_subtracted'].data[0][0])
    print(cal)
    return cal*counts
```
